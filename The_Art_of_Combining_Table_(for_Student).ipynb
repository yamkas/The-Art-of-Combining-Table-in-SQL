INNER JOIN

```python
#cek terlebih dahulu isi dari table customers
query = """
SELECT
* from customers
limit 5
"""
pd.read_sql_query(query,db)
```

```python
#cek terlebih dahulu isi dari table orders
query = """
SELECT
* from orders
limit 5
"""
pd.read_sql_query(query,db)
```

```python
#gabungkan company name, ordedate, orderid dan freight
query = """
SELECT
Customers.CompanyName,
date(Orders.OrderDate) as date_order,
Orders.OrderID,
orders.freight
FROM Customers
INNER JOIN Orders ON Customers.CustomerID = Orders.CustomerID
ORDER BY Customers.CompanyName
"""
pd.read_sql_query(query,db)
```

```python
#gabungkan company name, ordedate, orderid dan freight menggunakan alias
query = """
SELECT
a.CompanyName,
date(b.OrderDate) as date_order,
b.OrderID,
b.freight
FROM Customers a
INNER JOIN Orders b
ON a.CustomerID = b.CustomerID
ORDER BY 1
"""
pd.read_sql_query(query,db)
```

```python
#cari total order dan freight berdasarkan company name
query = """
SELECT
Customers.CompanyName,
count(distinct Orders.OrderID) as total_order,
sum(orders.freight) as freight
FROM Customers
INNER JOIN Orders ON Customers.CustomerID = Orders.CustomerID
group by 1
ORDER BY 1,2
"""
pd.read_sql_query(query,db)
```

```python
#cari total order dan freight berdasarkan company name dan bulan
query = """
SELECT
Customers.CompanyName,
date(Orders.OrderDate,'start of month') as date_order,
count(distinct Orders.OrderID) as total_order,
sum(orders.freight) as freight
FROM Customers
INNER JOIN Orders ON Customers.CustomerID = Orders.CustomerID
group by 1,2
ORDER BY 1,2
"""
pd.read_sql_query(query,db)
```

LEFT JOIN

```python
#cek kolom products
query = """
select * from products
"""
pd.read_sql_query(query, db)
```

```python
#cek kolom orderdetails
query = """
select * from orderdetails
limit 5
"""
pd.read_sql_query(query, db)
```

```python
#join product name dan order id dengan left join
query = """
SELECT
upper(a.productname) as product_name_upper,
b.orderid
FROM products a
LEFT JOIN Orderdetails b
on a.productid = b.productid
"""
pd.read_sql_query(query, db)
```

```python
#join product name dan order id dengan left join
query = """
SELECT
upper(productname) as product_name,
count(distinct b.orderid) as total_order
FROM products a
LEFT JOIN Orderdetails b
on a.productid = b.productid
group by 1
"""
pd.read_sql(query, db)
```

LEFT OUTER JOIN

```python
#join product name dan order id dengan left outer join
query = """
SELECT
upper(productname) as product_name,
count(distinct b.orderid) as total_order
FROM products a
LEFT JOIN Orderdetails b
on a.productid = b.productid
where b.productid is null
group by 1
"""
pd.read_sql(query, db)
```

RIGHT JOIN

```python
#join product name dan order id dengan left join namun ditukar tablenya untuk menghasilkan right join
query = """
SELECT
upper(productname),
count(distinct o.orderid)
FROM Orderdetails o
LEFT JOIN products p
on p.productid = o.productid
group by 1
"""
pd.read_sql(query, db)
```

UNION

```python
# menggabungkan companyname dari customers dan suppliers (TIPE DATANYA HARUS SAMA)
query = """
SELECT companyname from suppliers
union
SELECT companyname from customers
"""
pd.read_sql(query, db)
```

INTERSECT

```python
#mencari tau apakah semua employee pernah menjual pada tanggal tertentu
query = """
select employeeid from employees
intersect
select employeeid from orders
where date(OrderDate,'start of month') = '2014-08-01'
"""
pd.read_sql(query, db)
```

```python
#mencari tau apakah siapa employee tidak pernah menjual pada tanggal tertentu
query = """
select employeeid from employees
except
select employeeid from orders
where date(OrderDate,'start of month') = '2014-08-01'
"""
pd.read_sql(query, db)
```

1. Saya ingin mendapatkan total order, rata-rata freight dan detail karyawan yang menjual (nama depan, nama belakang, tittle, kota)


```python
query = """
SELECT
Employees.EmployeeID,
count(distinct Orders.OrderID) as total_order,
AVG(Orders.freight) as rata_rata_freight,
Employees.FirstName as nama_depan,
Employees.LastName as nama_belakang,
Employees.Title,
Employees.City as kota
FROM Employees
INNER JOIN Orders ON Employees.EmployeeID = Orders.EmployeeID
group by 1
ORDER BY 1,2
"""
pd.read_sql_query(query,db)
```

2. Berapa total penjualan (unitprice * quantity) dari  setiap kategori product?

```python
query = """
SELECT
Products.ProductID,
Products.ProductName,
SUM(OrderDetails.UnitPrice*OrderDetails.Quantity) as total_penjualan
FROM Products
LEFT JOIN OrderDetails ON OrderDetails.ProductID = Products.ProductID
GROUP BY 1
ORDER BY 3 ASC
"""
pd.read_sql_query(query,db)
```

3. 2. Berapa total penjualan (unitprice * quantity) dari  setiap kategori product pada tahun 2015?

```python
query = """
SELECT
Products.ProductID,
Products.ProductName,
SUM(OrderDetails.UnitPrice*OrderDetails.Quantity) as total_penjualan
FROM Products
LEFT JOIN OrderDetails ON Products.ProductID = OrderDetails.ProductID
LEFT JOIN Orders ON Orders.OrderID = OrderDetails.OrderID
WHERE date(Orders.OrderDate,'start of year') = '2015-01-01'
GROUP BY 1,2
ORDER BY 3
"""
pd.read_sql_query(query,db)
```
